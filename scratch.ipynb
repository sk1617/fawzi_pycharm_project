```python
import pandas as pd
from DataProcessing import main_data_processing, distance_formula
import plotly.graph_objects as go
outside_assignment_df = pd.read_table('/Volumes/Transcend/fawzi_pycharm_project/fto_assignment_files/Full_length/Run/assigned_peaks_res.out', sep=' +')
outside_peaks_df = pd.read_table('/Volumes/Transcend/fawzi_pycharm_project/fto_assignment_files/Full_length/CS17_FTO', sep=' +', index_col=False)
peak_list, residue_list = main_data_processing()

outside_peaks_df['N_'] = [float(x)+.23 for x in outside_peaks_df['N']]
outside_peaks_df['H_'] = [float(x)+.08 for x in outside_peaks_df['H']]
```

```python
a = 0
b = 0
for peak in peak_list:
    h_shift = peak.get_data('TROSYHShift')
    n_shift = peak.get_data('TROSYNShift')
    if h_shift is not None:
        ca_shift = peak.get_data('CAShift')
        cb_shift = peak.get_data('CBShift')
        ca_prime_shift = peak.get_data('CAPrimeShift')
        cb_prime_shift = peak.get_data('CBPrimeShift')
        # if ca_shift is None and cb_shift is None and ca_prime_shift is None and cb_prime_shift is None:
        #     a += 1
        #     print(outside_peaks_df[(abs(outside_peaks_df['N_'] - n_shift) < .5) & (abs(outside_peaks_df['H_'] - h_shift) < .1)])
        if ca_shift is None and ca_prime_shift is None:
            df = outside_peaks_df[(abs(outside_peaks_df['N_'] - n_shift) < .5) & (abs(outside_peaks_df['H_'] - h_shift) < .1)]
            if len(df) > 1:
                b += 1
                print(n_shift, h_shift, 'ca:', ca_shift, 'ca-1:', ca_prime_shift, 'cb:', cb_shift, 'cb-1:', cb_prime_shift)
                print(df)
                print('\n')
print(a, b)
```

```python
del_H_ls = list()
del_N_ls = list()
for i, peak in enumerate(peak_list):
    H_shift = peak.get_data('TROSYHShift')
    N_shift = peak.get_data('TROSYNShift')
    if H_shift is not None or N_shift is not None:
        HNCACBHShift = peak.get_data('HNCAHShift')
        HNCACBNShift = peak.get_data('HNCANShift')
        if HNCACBHShift is not None or HNCACBNShift is not None:
            delta_H = abs(H_shift - HNCACBHShift)
            delta_N = abs(N_shift - HNCACBNShift)
            del_H_ls.append(delta_H)
            del_N_ls.append(delta_N)
            print(delta_H, delta_N)
```
