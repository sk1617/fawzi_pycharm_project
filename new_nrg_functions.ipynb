```python
from EnergyFunctions import *
from DataImporting import *
import plotly.express as px
import plotly.graph_objects as go
import BMRB
```

```python
noesy_perfect_match_threshold = .02
npmt_penatly = 0.1
n_no_match_penalty = 600.

def dist_factor_forumla(dist):
    if dist < 3:
        return 2.8
    elif dist >= 3 and dist < 7:
        return 25/(dist ** 2)
    elif dist >= 7:
        return .5
def sn_factor_formula(sn):
    return (sn ** .5)

nearby_H_shift_signal_noise = 15
SN_factor = sn_factor_formula(nearby_H_shift_signal_noise)
dist_factor = dist_factor_forumla(5)
def noesy_old(delta): 
    if delta < noesy_perfect_match_threshold:
        subWeight = npmt_penatly
    else:
        subWeight = n_no_match_penalty * dist_factor * SN_factor
    return subWeight

x = [i/100 for i in range(150)]
y = [noesy_old(i) for i in x]
noesy_old_fig = px.scatter(x=x, y=y, title='NOESY Old')
noesy_old_fig.show()
```
